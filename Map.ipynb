```python
import requests
import pandas as pd
```

## Scraping Wikipedia and converting the table to dataframe

```python
website_url = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M").text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,"lxml")
```

```python
My_table = soup.find("table",{"class":"wikitable sortable"})
```

```python
df = pd.read_html(str(My_table))[0]
df
```

## Making first row values as column names of the table

```python
new_header = df.iloc[0] 
df = df[1:] 
df.columns = new_header
```

```python
df
```

## Dropping rows where Borough is 'Not assigned'

```python
indexNames = df[df['Borough'] == 'Not assigned'].index
df.drop(indexNames , inplace=True)
df
```

## Changing 'Not assigned' Neighbourhood values to the respective Borough value

```python
df['Neighbourhood'][df['Neighbourhood'] == 'Not assigned'] = df['Borough']
```

```python
df
```

## Combining the rows with the same postcodes (Neighbourhood names are separated by commas)

```python
df_group = df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(','.join).reset_index()
```

```python
df_group
```

## Exploring the neighbourhoods in Toronto, Canada

```python
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
!conda install -c conda-forge folium=0.5.0 --yes 
import folium
```

```python
address = 'Toronto, Canada'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronot, Canada are {}, {}.'.format(latitude, longitude))
```

```python
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_map['Latitude'], df_map['Longitude'], df_map['Borough'], df_map['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
